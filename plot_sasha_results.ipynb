```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
column_names=['q2','x','z','pt2','sumsibs/sumsib','sumsigsi/sumsigi','sumsigst/sumsigt','sumsibc/sumsib','sumsigci/sumsigi','sumsigct/sumsigt','sumsib','sumsigi','sumsigt','sumsibs','sumsigsi','sumsigst','sumsibc','sumsigci','sumsigct']
column_names=['q2','x','z','pt2','r1','r2','r3','r4','r5','r6','sumsib','sumsigi','sumsigt','sumsibs','sumsigsi','sumsigst','sumsibc','sumsigci','sumsigct']
filename = 'exc_compare/Tra-results/Exc/JLab2'
#filename = 'exc_compare/Tra-results/Exc/JLab3'
#filename = 'exc_compare/Tra-results/Exc/EIC1'
#filename = 'exc_compare/Tra-results/Exc/EIC2'
#filename = 'exc_compare/Tra-results/Exc/EIC3'
df_z1_sasha=pd.read_csv(filename,delim_whitespace=True,names=column_names)
df_z1_sasha.head()
```

```python
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r1'],color='black')
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r2'],color='blue')
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r3'],color='red')

plt.xlim(0,1)
plt.xlabel(r'$p_t^2(GeV^2)$')
plt.ylabel(r'$A_{UT}^{sin(\phi_h-\phi_s)}$')
plt.ylim(0,0.1)
plt.text(0.5, 0.02, f'$z: {df_z1_sasha["z"].iloc[0]}$', 
         horizontalalignment='center', verticalalignment='center')



plt.text(0.5, 0.1, f'$Q^2: {df_z1_sasha["q2"].iloc[0]}; x: {df_z1_sasha["x"].iloc[0]}$', 
         horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

```

```python
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r4'],color='black')
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r5'],color='blue')
plt.plot(df_z1_sasha['pt2'],df_z1_sasha['r6'],color='red')
plt.xlim(0,1)
plt.xlabel(r'$p_t^2(GeV^2)$')
plt.ylabel(r'$A_{UT}^{sin(\phi_h+\phi_s)}$')
plt.ylim(0,0.03)
plt.text(0.5, 0.01, f'$z: {df_z1_sasha["z"].iloc[0]}$', 
         horizontalalignment='center', verticalalignment='center')
plt.text(0.5, 0.15, f'$Q^2: {df_z1_sasha["q2"].iloc[0]}; x: {df_z1_sasha["x"].iloc[0]}$', 
         horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
```

```python
plt.plot(np.sqrt(df_z1_sasha['pt2']),df_z1_sasha['sumsib']/df_z1_sasha['sumsigt'],color='black',label='with Exc')
plt.plot(np.sqrt(df_z1_sasha['pt2']),df_z1_sasha['sumsib']/df_z1_sasha['sumsigi'],color='red',label='no Exc')

plt.xlim(0,1.2)
plt.xlabel(r'$P_{hT}(GeV/c)$')
plt.ylabel(r'RC')
plt.ylim(0.26,1.74)
#plt.text(0.5, 0.02, f'$z: {df_z1_sasha["z"].iloc[0]}$', 
#         horizontalalignment='center', verticalalignment='center')

plt.legend()

plt.text(0.5, 0.1, f'$Q^2: {df_z1_sasha["q2"].iloc[0]}; x: {df_z1_sasha["x"].iloc[0]}; z: {df_z1_sasha["z"].iloc[0]}$', 
         horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
plt.savefig(filename+'_RC.pdf')
```
