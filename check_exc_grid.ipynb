```python
def calculate_thcm(W, Q2, t):
    q0=(W**2-Q2-m_p**2)
    #print('q0',q0) 
    Ph0=(m_pi**2+W**2-m_n**2)
    numerator=q0*Ph0+2*W**2*(t+Q2-m_pi**2)
    denominator=np.sqrt(Ph0**2-4*m_pi**2*W**2)*np.sqrt(q0**2+4*Q2*W**2)
    thcm = math.acos(numerator / denominator)
    return 180*thcm/np.pi
```

```python
#file.open('exclu_t.grid','w')
for index, row in exclu.iterrows():
    W=row['W']/1000

    Q2=row['q^2']
    theta=row['theta']
    t=calculate_t_from_theta(W,Q2,theta)
    
    thetacm = calculate_thcm(W,Q2,t)
    if math.isnan(t):
        print('nan  W:%f,Q2:%f,theta:%f,t:%f,thcm:%f'%(W,Q2,theta,t,thetacm))
    else:
        print('W:%f,Q2:%f,theta:%f,t:%f,thcm:%f'%(W,Q2,theta,t,thetacm))
```
